```python
import pandas as pd
import math
```

```python
df = pd.read_excel("Task 1.xlsx" , sheet_name="Sheet1")
```

```python
dataset  = df.values.tolist()
header = df.columns.tolist()
```

```python
name = ""
count = 0
while True:
    if count>= len(dataset):
        break
    currentName = dataset[count][0]
#     print(currentName)
    if not pd.notnull(currentName):
        dataset[count][0] = name
    else:
        name = dataset[count][0]
        del dataset[count]
        count-=1
    count+=1
#     print(currentName,name)
```
