```python
import re
import pandas as pd
import calendar
from openpyxl import load_workbook
```

```python
df = pd.read_excel("Account Data Cleaned.xlsx")
```

```python
df
```

```python
df["Amount"] = df["Debit (AC)"] - df["Credit (RC)"]
```

```python
df.head(15)
```

```python
pattern = "(999|000).?0*$"
text = "0.00"
print(re.search(pattern,text))
```

```python
df["Amount"] = df["Amount"].astype(str)
```

```python
df[df["Amount"].str.contains(pattern)].sort_values("Amount")
```

```python
holidays = ["01-Jan","10-Apr","13-Apr","04-May","25-May","31-Aug","25-Dec","28-Dec"]
```

```python
from datetime import datetime
```

```python
SaturdayDF = pd.read_excel("SaturdayDF.xlsx")
SundayDF = pd.read_excel("SundayDF.xlsx")
HolidayDF = pd.read_excel("HolidayDF.xlsx")
for idx,i in enumerate(df["Date"]):
    if i.strftime("%A") == "Saturday":
        SaturdayDF = SaturdayDF.append(df.iloc[idx])
    if i.strftime("%A") == "Sunday":
        SundayDF = SundayDF.append(df.iloc[idx])
    day = i.strftime("%d")
    month = i.strftime("%b")
    hol = day + "-" + month
    if hol in holidays:
        HolidayDF = HolidayDF.append(df.iloc[idx])
```

```python
WeekendHolidaysDF = pd.DataFrame(columns= df.columns)
WeekendHolidaysDF = WeekendHolidaysDF.append(SaturdayDF)
WeekendHolidaysDF = WeekendHolidaysDF.append(SundayDF)
WeekendHolidaysDF = WeekendHolidaysDF.append(HolidayDF)
```

```python
WeekendHolidaysDF.reset_index(drop = True, inplace = True)
```

```python
writer = pd.ExcelWriter("Analysis.xlsx",engine="xlsxwriter")
WeekendHolidaysDF.to_excel(writer, sheet_name= "WeekendHolidays", index=False)
df[df["Amount"].str.contains(pattern)].sort_values("Amount").to_excel(writer,index = False, sheet_name = "Amount")
writer.save()
writer.close()
```
