```python
import csv
openFile = open("artworks.csv", encoding = "latin-1")
readFile = csv.reader(openFile)
dataset = list(readFile)
```

```python
header = dataset[0]
artwork = dataset[1:]
```

```python
for i in artwork: #NATIONALITY
    i[2] = i[2].replace("(","").replace(")","")
```

```python
for i in artwork: #GENDER
    i[5] = i[5].replace("(","").replace(")","")
```

```python
for i in artwork: #BEGIN DATE, END DATE
    if i[3] != "":
        i[3] = i[3].replace("-","")
        i[3] = int(i[3])
    if i[4] != "":
        i[4] = i[4].replace("-","")
        i[4] = int(i[4])
```

```python
badChars = ["C","c","(",")","."," ","'","s","S",]
```

```python
for i in artwork:
    for j in badChars:
        i[6] = i[6].replace(j,"")
    if i[6][0] == "-": #alternative is i[6].startswith("-")
        i[6] = i[6].replace("-","")
    if "-" in i[6]:
        year = i[6].split("-")
        i[6] = (int(year[0])+int(year[1]))/2
```

```python
for i in artwork:
    i[6] = int(i[6])
```

```python
def generateFreq(dataset, columnName):
    index = dataset[0].index(columnName)
    freq = {}
    for i in artwork:
        data = i[index].lower()
        if data == "":
            data = "Null"
        else:
            data = data.capitalize()
        if data in freq:
            freq[data]+=1
        else:
            freq[data] = 1
    return freq
```

```python
generateFreq(dataset,"Gender")
```

```python
print(generateFreq(dataset,"Nationality"))
```
