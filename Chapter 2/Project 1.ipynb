```python
import csv
file = open ("googleplaystore.csv",encoding = "UTF-8")
read = csv.reader(file)
dataset = list(read)
```

```python
header = dataset[0]
android = dataset[1:]
```

```python
def checkEmpty(datasets, columnName):
    index = datasets[0].index(columnName)
#     print(index)
    emptyCells = []
    for i in datasets[1:]:
        if i[index] == "":
            emptyCells.append(datasets.index(i))
    return emptyCells
```

```python
print(checkEmpty(dataset,"Content Rating"))
```

```python
duplicateApp = []
uniqueApp = []
for i in android:
    app=i[0]
    if app in uniqueApp:
        duplicateApp.append(app)
    else:
        uniqueApp.append(app)
```

```python
reviews = {}
for i in android:
    appName = i[0]
    appReviews = int(i[3])
    if appName not in reviews:
        reviews[appName] = appReviews
    else:
        reviews[appName] = max(reviews[appName],appReviews)
```

```python
reviews["Instagram"]
```

```python
androidClean = []
addedList = []
for i in android:
    appReviews = int(i[3])
    appName = i[0]
    if reviews[appName] == appReviews and appName not in addedList:
        androidClean.append(i)
        addedList.append(i[0])
androidClean
```

```python
def isEnglish(appName):
    count = 0
    for i  in appName:
        if ord(i)>127 and count>=3:
            return False
        if ord(i)>127:
            count+=1
    return True
```

```python
i = 0
while(True):
    if(i>=len(androidClean)):
        break
    name  = androidClean[i][0]
    if not isEnglish(name):
        androidClean.remove(androidClean[i])
        i-=1
    i+=1
```

```python
categoryAnalysis = []

for i in uniqueCategories:
    totalInstalls = 0
    numberOfApps = 0
    for j in freeApps:
        if j[1]==i:
            installs = j[5]
            totalInstalls += int(installs.replace('+','').replace(',',''))
            numberOfApps+=1
    categoryAnalysis.append([round(totalInstalls/numberOfApps),i])
```
