```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.espncricinfo.com/rankings/content/page/211271.html")
```

```python
response
```

```python
content = response.text
```

```python
content
```

```python
soup = BeautifulSoup(content,'html.parser')
```

```python
soup
```

```python
def getTables(table,index):
    tableData = table[index]
    td = tableData.find_all('td')
    tableVals = []
    for i in range(len(td)//5):
        data = td[5*i:5*(i+1)]
        data = [i.text for i in data]
        tableVals.append(data)
    return tableVals
```

```python
head = soup.find_all("th")
head = head[0:5]
head = [i.text for i in head]
writer = pd.ExcelWriter("CricInfo.xlsx",engine = 'xlsxwriter')
tables = soup.find_all('table',{'class':'StoryengineTable'})
tableFormat = soup.find_all('h3')
tableFormat = [i.text for i in tableFormat]
for i in range(len(tableFormat)):
    data = getTables(tables,i)
    df = pd.DataFrame(data=data,columns=head)
    df.to_excel(writer,sheet_name=tableFormat[i],index=False)
writer.save()
```
